# With Loop

```python
df = pd.read_csv('supershop.csv')
df.head()
```

```python
cols = ['Area']

for col in cols:
    one_hot = pd.get_dummies(df[col])
    df = pd.concat((df,one_hot),axis=1).drop(col,axis=1)
```

```python
df.head()
```

```python
df = pd.read_csv('supershop.csv')
df.head()
```

```python
cols = ['Area']

for col in cols:
    one_hot = pd.get_dummies(df[col],drop_first=True,prefix=col)
    df = pd.concat((df,one_hot),axis=1).drop(col,axis=1)
```

```python
df.head()
```

# Ordinal Encoder

```python
df = pd.read_csv('supershop.csv')
df.head()
```

```python
df.Area.unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
city = ['Dhaka', 'Ctg', 'Rangpur']
```

```python
ordinal = OrdinalEncoder(categories=[city])
```

```python
encodedValue = ordinal.fit_transform(df[['Area']])
```

```python
encodedValue
```

```python
new_area = pd.DataFrame(encodedValue,columns=['Area'])
```
