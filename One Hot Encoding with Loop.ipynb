```python
import pandas as pd
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
col = ['sex','smoker','region']

for column in col:
    one = pd.get_dummies(df[column], drop_first=True,prefix='dummy')
    df = pd.concat([df,one],axis=1).drop(column,axis=1)
    
```
